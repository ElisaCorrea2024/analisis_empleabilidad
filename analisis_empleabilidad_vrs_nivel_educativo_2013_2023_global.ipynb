```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el dataset limpio
df = pd.read_csv('ocde_data_education_cleaned.csv')

# Paso 1: Calcular la tasa promedio de empleo por cada nivel educativo
empleabilidad_por_nivel = df.groupby('Education Level')['Value'].mean().sort_values(ascending=False)

# Paso 2: Visualizar los resultados en un grfico de barras
plt.figure(figsize=(10, 6))
empleabilidad_por_nivel.plot(kind='bar', color='skyblue')
plt.title('Tasa Promedio de Empleo por Nivel Educativo')
plt.xlabel('Nivel Educativo')
plt.ylabel('Tasa de Empleo Promedio')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Exportar los resultados del anlisis de empleabilidad por nivel educativo a un archivo CSV
empleabilidad_por_nivel.to_csv('empleabilidad_por_nivel_educativo.csv', header=True)

print("Anlisis exportado como 'empleabilidad_por_nivel_educativo.csv'")
```
