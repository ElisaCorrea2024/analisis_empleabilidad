```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el archivo CSV limpio y actualizado con el nombre
df = pd.read_csv('ocde_data_nivel_de_educacion_y_edades_vr_empleabilidad_cleaned.csv')

# Filtrar el ao ms reciente y el rango de edad 25-34 aos
year_most_recent = df['Year'].max()
df_25_34 = df[(df['Year'] == year_most_recent) & (df['Age'] == '25-34 aos')]

# Agrupar por 'Education Level' y calcular el promedio de empleabilidad para el grupo de 25-34 aos
empleabilidad_25_34 = df_25_34.groupby('Education Level')['Value'].mean().reset_index()

# Visualizar los resultados en un grfico de barras
plt.figure(figsize=(10, 6))
sns.barplot(data=empleabilidad_25_34, x='Education Level', y='Value', color='skyblue')
plt.title(f'Tasa Promedio de Empleo por Nivel Educativo para el Grupo de 25-34 aos - {year_most_recent}')
plt.xlabel('Nivel Educativo')
plt.ylabel('Tasa de Empleo Promedio')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Exportar el anlisis a un archivo CSV con el nombre especificado
empleabilidad_25_34.to_csv('analisis_empleabilidad_edades_25_a_35_aos_2023_global.csv', index=False)
print("Anlisis exportado como 'analisis_empleabilidad_edades_25_a_35_aos_2023_global.csv'")
```
